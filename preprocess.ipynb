```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings
import os
from dotenv import load_dotenv
warnings.filterwarnings('ignore')
```

```python
#loading data
import kagglehub

# Download latest version
path = kagglehub.dataset_download("firecastrl/us-wildfire-dataset")
print(path)
```

```python
load_dotenv(override=True)
data_path = os.getenv("DATA_PATH")
df = pd.read_csv(data_path)
```

```python
# Convert Wildfire to binary
df['Wildfire_binary'] = (df['Wildfire'] == 'Yes').astype('int8')

print("Wildfire distribution:")
print(df['Wildfire_binary'].value_counts())
print(f"Wildfire rate: {df['Wildfire_binary'].mean():.3f}")
```

```python
# Simple approach for days_from_ignition
print("\nCreating temporal sequences...")

# Reset index to ensure proper ordering
df = df.reset_index(drop=True)

# Create sequence groups (each 75-day window)
df['window_id'] = df.index // 75
df['days_from_ignition'] = (df.index % 75) - 60

print(f"Total windows: {df['window_id'].nunique()}")
print(f"Days from ignition range: {df['days_from_ignition'].min()} to {df['days_from_ignition'].max()}")
print(f"Window size check: {df['window_id'].value_counts().head(3)}")

#there are 126798 unique 75-day wildfire windows
```

```python
#normalization with z score
normalization_cols = ['pr', 'rmax', 'rmin', 'sph', 'srad', 'tmmn', 'tmmx', 'vs', 
                     'bi', 'fm100', 'fm1000', 'erc', 'etr', 'pet', 'vpd']

scaler = StandardScaler()
df[normalization_cols] = scaler.fit_transform(df[normalization_cols])
```
